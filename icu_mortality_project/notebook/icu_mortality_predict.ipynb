Dropping Features that are too missing

```python
y = df["mortality"]
```

```python
y
```

```python
df.columns
```

```python
X = df.drop(columns=[
    'mortality','subject_id', 'hadm_id','stay_id','intime','outtime','los'
])
```

```python
X
```

Encoding Categorical Variables

```python
X["gender"] = X["gender"].map({'M': 1, 'F':0})
```

```python
X
```

```python

X["first_careunit"].value_counts().sort_index()
```

```python
X = pd.get_dummies(
    X,
    columns=["first_careunit"],
    drop_first=True
)
```

```python
X = X.drop(columns=['last_careunit'])
```

```python
X
```

Impute missing values

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
X_imputed = imputer.fit_transform(X)
```

```python
X_imputed
```

```python
X_imputed
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_imputed,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score

y_pred_proba = model.predict_proba(X_test)[:, 1]
auc = roc_auc_score(y_test, y_pred_proba)

print("Baseline AUC:", auc)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
model.fit(X_train, y_train)

y_pred_proba = model.predict_proba(X_test)[:, 1]
auc = roc_auc_score(y_test, y_pred_proba)
print("Improved AUC:", auc)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X_train, y_train, cv=5, scoring='roc_auc')
print("CV AUC:", scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]
}

rf = RandomForestClassifier(random_state=42)
```

```python
grid_rf = GridSearchCV(
    rf,
    param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)

grid_rf.fit(X_train, y_train)
```

```python
print("Best CV AUC:", grid_rf.best_score_)
print("Best Parameters:", grid_rf.best_params_)
```

Trainning the final Random Forest Using the best parameters.

```python
best_params = grid_rf.best_params_

final_rf = RandomForestClassifier(
    **best_params,
    random_state=24
)

final_rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score

y_pred_proba = final_rf.predict_proba(X_test)[:,1]
auc_test = roc_auc_score(y_test, y_pred_proba)
print("Test set AUC:", auc_test)
```
